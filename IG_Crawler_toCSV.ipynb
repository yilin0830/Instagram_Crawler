```python
hashtag_list = ['yunlin']
hashtag = hashtag_list[0]

# 開啟輸出的 CSV 檔案
with open(hashtag + '.csv', 'w', encoding="utf-8", newline='') as csvfile:
    # 建立 CSV 檔寫入器
    writer = csv.writer(csvfile)
    # 寫入第一列資料
    writer.writerow(['comments_disabled', '__typename', '__id', 'text', 'shortcode', 'link', 'edge_media_to_comment', 'taken_at_timestamp', 'display_url', 'edge_liked_by', 'edge_media_preview_like', 'owner', 'is_video', 'accessibility_caption'])
    
    
    url = 'https://www.instagram.com/explore/tags/' + hashtag +'/?__a=1'
    html = urlopen(url).read()
    soup = bs(html,"html.parser")
    page_json = soup.text.rstrip(';')
    data = json.loads(page_json)
    
    post_num = data['graphql']['hashtag']['edge_hashtag_to_media']['count'] # 總貼文數量
    for post in data['graphql']['hashtag']['edge_hashtag_to_media']['edges']:
        comments_disabled =  post['node']['comments_disabled']
        __typename = post['node']['__typename']
        __id = post['node']['id']
        try:
            text = post['node']['edge_media_to_caption']['edges'][0]['node']['text']
        except:
            text = ''
        shortcode = post['node']['shortcode']
        link ='https://www.instagram.com'+'/p/'+ post['node']['shortcode']+'/'  
        edge_media_to_comment = post['node']['edge_media_to_comment']['count']
        taken_at_timestamp = post['node']['taken_at_timestamp']
        display_url = post['node']['display_url']
        edge_liked_by = post['node']['edge_liked_by']['count']
        edge_media_preview_like = post['node']['edge_media_preview_like']['count']
        owner = post['node']['owner']['id']
        is_video =  post['node']['is_video']
        accessibility_caption = post['node']['accessibility_caption']

        writer.writerow([comments_disabled, __typename, __id, text, shortcode, link, edge_media_to_comment, taken_at_timestamp, display_url, edge_liked_by, edge_media_preview_like, owner, is_video, accessibility_caption])
    
    has_next_page = data['graphql']['hashtag']['edge_hashtag_to_media']['page_info']['has_next_page']
    while has_next_page:     
        end_cursor = data['graphql']['hashtag']['edge_hashtag_to_media']['page_info']['end_cursor']
        next_page = 'https://www.instagram.com/explore/tags/'+ hashtag +'/?__a=1&max_id=' + end_cursor
        url = next_page

        html = urlopen(url).read()
        soup = bs(html,"html.parser")
        page_json = soup.text.rstrip(';')
        data = json.loads(page_json)
        for post in data['graphql']['hashtag']['edge_hashtag_to_media']['edges']:
            comments_disabled =  post['node']['comments_disabled']
            __typename = post['node']['__typename']
            __id = post['node']['id']
            
            try:
                text = post['node']['edge_media_to_caption']['edges'][0]['node']['text']
            except:
                text = ''
            
            shortcode = post['node']['shortcode']
            link ='https://www.instagram.com'+'/p/'+ post['node']['shortcode']+'/'  
            edge_media_to_comment = post['node']['edge_media_to_comment']['count']
            taken_at_timestamp = post['node']['taken_at_timestamp']
            display_url = post['node']['display_url']
            edge_liked_by = post['node']['edge_liked_by']['count']
            edge_media_preview_like = post['node']['edge_media_preview_like']['count']
            owner = post['node']['owner']['id'] 
            is_video =  post['node']['is_video']
            accessibility_caption = post['node']['accessibility_caption']

            writer.writerow([comments_disabled, __typename, __id, text, shortcode, link, edge_media_to_comment, taken_at_timestamp, display_url, edge_liked_by, edge_media_preview_like, owner, is_video, accessibility_caption])
        
        has_next_page = data['graphql']['hashtag']['edge_hashtag_to_media']['page_info']['has_next_page']
```
