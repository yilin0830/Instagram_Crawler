```python
import csv
hashtag = 'chinesetaipei'
texts = []
with open(hashtag + '.csv', encoding='UTF-8') as csvfile:

    # 讀取 CSV 檔案內容
    rows = csv.DictReader(csvfile)

    # 以迴圈輸出每一列
    for row in rows:
        texts.append(row['text'])
       
```

```python
import re
seg_list= list()
for i in range(len(texts)):
    hashtags = re.findall(r"#(\w+)", texts[i])
    for i in hashtags:
        seg_list.append(i) 
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import imageio

wc = WordCloud(
    background_color='white',
    collocations=False,
    mask=imageio.imread('./WordCloud/Taiwan.jpg'),
    random_state=10,
    font_path='./WordCloud/kaiu.ttf'
                ).generate(' '.join(seg_list))
plt.imshow(wc)
plt.axis('off')
plt.show()

wc.to_file('./WordCloud/'+hashtag+ '.png')
```
